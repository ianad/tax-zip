```python
import os
import pandas as pd
from datascience import Table
```

```python
data = pd.read_csv('zipcode-data-2015.csv')
```

```python
descriptions = pd.read_csv('zipcode-data-descriptions.csv')
```

```python
columns = list(descriptions['VARIABLE NAME'])
new_columns = list(descriptions['DESCRIPTION'])
new_columns = [s.replace(' ','-').lower() for s in new_columns]
new_columns[0:3] = columns[0:3]
new_columns = [s.strip(' ').lower() for s in new_columns]
```

```python
data.columns = new_columns
```
